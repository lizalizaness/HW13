```python
!pip install numpy
!pip install pandas
!pip install matplotlib
!pip install statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
income = np.array([1800, 12032, 1500, 1344, 5200, 2800, 53567, 4739, 54658, 2000])
income_without_tax = income * 0.7
expenses = np.array([800, 802, 1000, 95, 1800, 1900, 980, 1000, 1200, 1300])

savings = income_without_tax - expenses
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]
df = pd.DataFrame({
    'Income Without Tax': income_without_tax,
    'Expenses': expenses,
    'Savings': savings}, index=months)

plt.figure(figsize=(10, 5))
plt.plot(months, df['Income Without Tax'], label="Income Without Tax", color='blue', marker='o')
plt.title('Income by Month')
plt.xlabel('Month')
plt.ylabel('Income Without Tax')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(months, df['Savings'], color='green')
plt.title('Savings by Month')
plt.xlabel('Month')
plt.ylabel('Savings')
plt.show()

plt.figure(figsize=(7, 7))
plt.pie(df['Savings'], labels=months, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Monthly Savings as a Percentage of Total Savings')
plt.show()

from statistics import mean
quarters = {
    'Q1': df.loc[['Jan', 'Feb', 'Mar'], 'Income Without Tax'],
    'Q2': df.loc[['Apr', 'May', 'Jun'], 'Income Without Tax'],
    'Q3': df.loc[['Jul', 'Aug', 'Sep'], 'Income Without Tax'],
    'Q4': df.loc[['Oct'], 'Income Without Tax']
}

average_income_per_quarter = {q: mean(quarters[q]) for q in quarters}
print("Average Income per Quarter:")
for q, avg_income in average_income_per_quarter.items():
    print(f"{q}: {avg_income}")
```

```python
!pip install seaborn
!pip install matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

df_car_crashes = sns.load_dataset('car_crashes')
sns.pairplot(df_car_crashes)
plt.title('Pairplot of Car Crashes Dataset')
plt.show()
numeric_df_car_crashes = df_car_crashes.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df_car_crashes.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Car Crashes Dataset (Numeric Data Only)')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x='ins_premium', y='total', data=df_car_crashes, palette='Blues')
plt.title('Total Crashes by Insurance Premiums')
plt.show()
```

```python
def cubic_algo(n):
    count = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                count += 1
    print(count)

cubic_algo(5)
```
